### Data Cleaning
Data cleaning means fixing bad data in your data set.

Bad data could be:

- Empty cells
- Data in wrong format
- Wrong data
- Duplicates

```python
#cleaning empty cells
# reading json file
df = pd.read_json('data.json')
df_copy = df.copy()
new_df = df_copy.dropna()

print("Initial shape", df_copy.shape, sep="\n")
print("After removed empy rows", new_df.shape,sep="\n")
```

```python
## replace empty values
replaced_df = df_copy.fillna(0)
print(df_copy[:10])
print(replaced_df[:10])
```

```python
print(df_copy.head(20))
```

```python
x = df_copy["Calories"].mean()
df_copy["Calories"].fillna(x, inplace=True)
print(df_copy.head(20))
```

### Cleaning Data of Wrong Format
Cells with data of wrong format can make it difficult, or even impossible, to analyze data.

To fix it, you have two options: remove the rows, or convert all cells in the columns into the same format.

```python
df.head(30)
```

### Fixing Wrong Data
If you take a look at our data set, you can see that in row 7, the duration is 450, but for all the other rows the duration is between 30 and 60.

```python
 # replacing the values 
df.loc[7, "Duration"] = 45

# setting bondaries for legal values
for x in df.index:
    if df.loc[x, "Duration"] > 120:
        df.loc[x, "Duration"] = 120

#  removing rows not passing a certain boundary
for x in df.index:
    if df.loc[x, "Duration"] > 120:
        df.drop(x, inplace=True)
```

### Removing Duplicates

```python
print(df.duplicated())
df.drop_duplicates(inplace=True)
df.head(30)
```
