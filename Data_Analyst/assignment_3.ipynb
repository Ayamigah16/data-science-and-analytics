```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline
```

### Task 1
The	Sklearn	library	has	a	built-in	dataset,	Iris.	It	contains	samples	from	various	species	of	the	iris	plant.	Each
sample	contains	four	attributes:	sepal	length,	sepal	width,	petal	length,	petal	width,	and	species	(Iris	setosa,
Iris	versicolor,	and	Iris	virginica),	and	there	are	50	samples	of	each	species.

1. Read	the	data	from	this	dataset	into	a	DataFrame.
2. Create	a	10*10  figure with two subplots
3. In the first subplot plot the sepal 
4. In the second subplot, plot the sepal length vs sepal width
5. For each of the plots, label the x and y  axes and set a title

```python
# loading the datasets
from sklearn.datasets import load_iris
iris_dataset = load_iris()  
iris_data = pd.DataFrame(data=iris_dataset.data, columns=iris_dataset.feature_names)
iris_data.head(10)
```

```python
# creating a 10*10 figure
fig = plt.figure(figsize=(3, 10))

# the first subplot
ax1 = fig.add_subplot(2,1,1)
ax1.scatter(iris_data["petal length (cm)"],iris_data["petal width (cm)"])

# setting the labels and the title
ax1.set_xlabel("Petal Length (cm)")
ax1.set_ylabel("Petal Width (cm)")
ax1.set_title("Petal Length (cm) Against Petal Width (cm)")

# the seccond subplot
ax2 = fig.add_subplot(2,1,2)
ax2.scatter(iris_data["sepal length (cm)"], iris_data["sepal width (cm)"])

# setting the labels and the title
ax2.set_xlabel("Sepal Length (cm)")
ax2.set_ylabel("Sepal Width (cm)")
ax2.set_title("Sepal Length (cm) Against Sepal Width (cm)")

plt.suptitle("IRIS DATASETS")
```

## Task 2
Load	the	data	from	tips	dataset	(built	into	Seaborn)	using	the	load_dataset	function	in	Seaborn.	This	dataset
contains	the	total	bill	amounts	and	the	tip	values	for	different	customers	visiting	a	restaurant.	The
customers	are	categorized	according	to	their	gender,	smoking	preference,	and	the	day	and	time	of	their	visit.
1. Create	a	plot	that	shows	the	distribution	(strip	plot)	of	the	total	bill	for	smokers	and	nonsmokers,	across	a
grid	containing	different	values	for	the	time	and	sex	columns.
2. Create	a	plot	to	show	the	relationship	between	the	“tip”	and	“total_bill”	columns	for:	males	and	smokers,
males	and	nonsmokers,	females	and	smokers,	and	females	and	nonsmokers.


```python

# loading the tips dataset
tips_dataset = sns.load_dataset("tips")
```

```python
tips_dataset.head()
```

```python
tips_dataset.columns
```

```python
""" 
creating a catplot for showing the distribution of the total bill for
different combinations of parameters
"""
sns.catplot(x='smoker', y="total_bill", row="time", col="sex", data=tips_dataset)
```

```python
""" 
#defining the FacetGrid object and setting the row and column values
"""
grid = sns.FacetGrid(data=tips_dataset, row="sex", col="smoker")
grid.map(plt.scatter, 'tip', 'total_bill')
```
