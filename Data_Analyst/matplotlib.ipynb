```python
import matplotlib.pyplot as plt  # a module for ploting graphs
import numpy as np

# importing math module to access some trig functions
import math
```

```python
def sin_angle(x):
    """ 
    Computes the sine of an angle

    Parameter:
    x = the angle to compute its sine

    Return:
    The sine of angle x
    """
    angle_to_radian = math.radians(x)
    return math.sin(angle_to_radian).__round__(2)

def cos_angle(x):
    """ 
    Computes the cosine of an angle

    Parameter:
    x = the angle to compute its cosine

    Return:
    The sine of angle x
    """
    angle_to_radian = math.radians(x)
    return math.cos(angle_to_radian).__round__(2)
```

```python
print(sin_angle(30), cos_angle(60), sep="\n")
```

## Displaying Multiple Plots Using "subplot()"

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

#the figure has 1 row, 2 columns, and this plot is the first plot.
plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

#the figure has 1 row, 2 columns, and this plot is the second plot.
plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

### Compare Plots
In the example above, there seems to be a relationship between speed and age, but what if we plot the observations from another day as well? Will the scatter plot tell us something else?

```python
# Comparing Plots

#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, c="black")

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, c="red")

plt.xlabel("Age Of Cars")
plt.ylabel("Speed of Cars")

plt.show()
```

## ColorMap
A colormap is like a list of colors, where each color has a value that ranges from 0 to 100.

#### Some colormaps
- Accent
- Blues
- CMRmap
- Greys
- twillight
- winter
- terrain
- viridis, etc

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

# adding colors and sizes
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])


plt.scatter(x, y, c=colors, s = sizes, alpha = 0.5, cmap='winter')

# adding the colormap bar to the drawing
plt.colorbar()

plt.show()
```

### Matplotlib Histograms Using "hist()"
A histogram is a graph showing frequency distributions.

It is a graph showing the number of observations within each given interval.


```python
# randomly generate an array with 250 values, where the values will concentrate around 170
x = np.random.normal(170, 10, 250)
plt.title("Histogram")
plt.hist(x)
plt.show()
```
