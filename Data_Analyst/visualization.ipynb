## Commonly Used Plots
* **Bar chart :** A bar chart enables visualization of categorical data, with the width or height of the bar representing the value for
each category. The bars can be shown either vertically or horizontally.

* **Histogram :**  A histogram is used to visualize the distribution of a continuous variable. It divides the range of the continuous variable into intervals and shows where most of the values lie

* **Box plots:** Box plots help with visually depicting the statistical characteristics of the data. A box plot provides a five
point summary with each line in the figure representing a statistical measure of the data being plotted.

* **Pie Charts :**  A pie chart shows the distinct values of a variable as sectors within a circle. Pie charts are used with categorical variables.

* **Scatter Plots :** A scatter plot displays the values of two continuous variables as points on the x and y axes and helps us visualize if the two variables are correlated or not.

* **Heat Maps :**  A heat map shows the correlation between multiple variables using a color-coded matrix, where the color
saturation represents the strength of the correlation between the variables. A heat map can aid in the visualization of multiple variables at once.

### Interfaces in Maatplotlib

1. ***Stateful Interface :*** 
This interface uses the pyplot class, which is based on Matlab.
A single object of this class is created, which is used for all
plotting purposes.

2. ***Object Oriented Interface :***
ace, we use different objects for different elements of the plot. The two main objects that are used in this interface for plotting are
• The figure object, which acts as the actual container for other objects.
• The axes object, which is the actual plotting area containing the x-axis,  y-axis, points, lines, legends and labels.


```python
# Stateful Interface

import matplotlib.pyplot as plt
%matplotlib inline 

x = [1, 2, 3, 4]
y = [3, 6, 9, 12]

plt.plot(x,y) # plots the line graph

# stting the x and y limits
plt.xlim(0, 6)
plt.ylim(0, 18)

# setting the labels
plt.xlabel("Events")
plt.ylabel("Outcomes")

plt.title("Today")

plt.suptitle("Main Title")  # set the main title
plt.show()
```

```python
#   Object Oriented Interface

x = [1, 2, 3, 4]
y = [3, 6, 9, 12]

# creating the plot objects
fig, ax = plt.subplots(figsize=(5, 5))

ax.plot(x, y)
ax.set_xlabel("Events")
ax.set_ylabel("Outcomes")

ax.set_title("Today")


plt.show()
```

```python
fig = plt.figure(figsize=(5, 12))
plt.suptitle("Overall")

# creating two subplots
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2,1,2)

# setting up the subplots area
ax1.set_title("Histogram")
ax1.set_xlabel("Age")
ax1.set_ylabel("Frequency")

ax2.set_xlabel("Gender Category")
ax2.set_ylabel("Number")
ax2.set_title("Bar Graph")

import pandas as pd
data = pd.read_csv("titanic.csv")
# plotting the histogram
ax1.hist(data["Age"])

# plotting the bar graph with the Sexes
category = ["Male", "female"]
values = [577, 314]
ax2.bar(category, values)

plt.show()
```

```python
import seaborn as sns 
titanic_data =pd.read_csv("titanic.csv")
titanic_data.head()
```

```python
# boxplots
sns.boxenplot(titanic_data["Age"])
```

```python
# kernel density estimate
# a	plot for visualizing the probability distribution of a continuous variable
sns.kdeplot(titanic_data["Age"])
```

```python
# violin plot
# 	merges	the	box	plot	with	the	kernel	density	plot,	with	the	shape	of	the	violin	representing	the
# frequency	distribution
sns.violinplot(x="Pclass", y="Age", data=titanic_data)
```

```python
# COUNT PLOTS
""" 
used	to	plot	categorical	variables,	with	the	length	of	the	bars	representing	the	number	of
observations	for	each	unique	value	of	the	variable
"""
sns.countplot(titanic_data['Survived'])
```

```python
titanic_data['Survived'].describe
```

```python
# heat maps
""" 
A	heatmap	is	a	graphical	representation	of	a	correlation	matrix,	representing	the	correlation	between
different	variables	in	a	dataset. The	intensity	of	the	color	represents	the	strength	of
the	correlation,	and	the	values	represent	the	degree	of	correlation
"""
fig = plt.figure(figsize=(20,20))

features = ["PassengerId", "Survived", "Pclass", "Age","SibSp", "Parch", "Fare"]
sns.heatmap(titanic_data[features], cmap="YlGnBu", robust=True, vmax=1, vmin=0)
plt.show()
```

```python
# facet grid
""" 
A	facet	grid	represents	the	distribution	of	a	single	parameter	or	the	relationship	between	parameters	across
a	grid	containing	a	row,	column,	or	hue	parameter
"""
# creating a grid object
grid = sns.FacetGrid(titanic_data, row="Survived", col="Sex")

# using the grid object to plot a graph of our choice
grid.map(plt.hist, "Age")
```

```python
# Reg Plot[Line of Best Fit]
""" 
This	plot	uses	the	linear	regression	model	to	plot	a	regression	line	between	the	data	points	of	two
continuous	variables
"""
sns.regplot(data=titanic_data, x="Age", y="Fare")
```

```python
# Implot > Reg Plot + Facet Plot Grid
sns.lmplot(x="Age", y="Fare", row="Survived", data=titanic_data)
```

```python
# Strip Plot
""" 
A	strip	plot	is	similar	to	a	scatter	plot.	The	difference	lies	in	the	type	of	variables	used	in	a	strip	plot.	While	a
scatter	plot	has	both	variables	as	continuous,	a	strip	plot	plots	one	categorical	variable	against	one
continuous	variable
"""
sns.stripplot(x="Survived", y="Age", data=titanic_data)
```

```python
# Swarm Plot
""" 
A	swarm	plot	is	similar	to	a	strip	plot,	the	difference	being	that	the	points	in	a	swarm	plot	are	not
overlapping	like	those	in	a	strip	plot.	With	the	points	more	spread	out,	we	get	a	better	idea	of	the
distribution	of	the	continuous	variable
"""
sns.swarmplot(x="Survived", y="Age", data=titanic_data)
```

```python
# Cat plot >> Strip Plot + Facet Grid
sns.catplot(x="Survived", y="Age", col="Survived", row="Sex", data=titanic_data)
```

```python
# Paair Plot
""" 
A	pair	plot	is	one	that	shows	bivariate	relationships	between	all	possible	pairs	of	variables	in	the	dataset.
Notice	that	you	do	not	have	to
supply	any	column	names	as	arguments	since	all	the	variables	in	the	dataset	are	considered	automatically
for	plotting.	The	only	parameter	that	you	need	to	pass	is	the	name	of	the	DataFrame
"""
sns.pairplot(titanic_data)
```

```python
# Joint Plot
""" 
The	joint	plot	displays	the	relationship	between	two	variables	as	well	as	the	individual	distribution	of	the
variables.
The	jointplot	function	takes	the	names	of	the	two	variables	to	be	plotted
as	arguments.
"""
sns.jointplot(x="Fare", y="Age", data=titanic_data)
```

## Which	plot	would	you	use	when	you	have	the	following	variables
1. One	categorical	variable:	count	plot
2. One	continuous	variable:	histogram,	kernel	density	estimate
3. Two	continuous	variables:	scatter	plot,	line	plot
4. Three	or	more	continuous	variables:	heat	map
5. One	continuous	and	one	categorical	variable:	strip	plot,	swarm	plot,	bar	plot
6. Two	continuous	and	two	or	more	categorical	variables:	catplot,	facet	grid,	lmplot
